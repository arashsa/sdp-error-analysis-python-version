```python
import pygal
from pygal.style import CleanStyle
import cairosvg
import lxml, tinycss, cssselect
```

```python
def graf_dependencies(results, f, to, name):
    bar_chart = pygal.Bar(legend_at_bottom=True, 
                          legend_at_bottom_columns=3, 
                          logarithmic=True, 
                          style=CleanStyle, 
                          x_label_rotation=30)
    
    bar_chart.x_labels = results[0][f:to]
    gold = results[1]
    pred = results[2]
    match = results[3]

    bar_chart.add('Gold', gold[f:to])
    bar_chart.add('Predicted', pred[f:to])
    bar_chart.add('Match', match[f:to])
    bar_chart.render_to_file('plots/dependencies/{}_dependencies.svg'.format(name))
```

```python
graf_dependencies(lisbon_plot1, 0, 20, 'Lisbon_DM')
graf_dependencies(peking_plot1, 0, 20, 'Peking_DM')
graf_dependencies(turku_plot1, 0, 20, 'Turku_DM')

graf_dependencies(lisbon_plot2, 0, 20, 'Lisbon_PSD')
graf_dependencies(peking_plot2, 0, 20, 'Peking_PSD')
graf_dependencies(turku_plot2, 0, 20, 'Turku_PSD')
```

```python
def graf_prf(results, f, to, name, parsers, p_r_f):
#     chart = pygal.Line(legend_at_bottom=True, legend_at_bottom_columns=3, logarithmic=True, style=CleanStyle)
    chart = pygal.Bar(range=(.2, 1),
                      legend_at_bottom=True, 
                      legend_at_bottom_columns=3, 
                      x_label_rotation=30, 
                      style=CleanStyle)
    
    chart.x_labels = results[0][0][f:to]
    name_p_r_f = {'p': 'Precision', 'r': 'Recall', 'f': 'F-score'}
#     chart.title = '{} {}'.format(name.replace('_', ' '), name_p_r_f[p_r_f])
    
    for i in range(3):
        gold1 = results[i][4]
        pred1 = results[i][5]
        match1 = results[i][6]

        if p_r_f == 'p':
            chart.add('{}'.format(parsers[i]), gold1[f:to])
        elif p_r_f == 'r':
            chart.add('{}'.format(parsers[i]), pred1[f:to])
        elif p_r_f == 'f':
            chart.add('{}'.format(parsers[i]), match1[f:to])
        
    chart.render_to_file('plots/dependencies/{}_{}_dependencies.svg'.format(name.replace('_', ' '), name_p_r_f[p_r_f]))
```

```python
graf_prf((lisbon_plot1, peking_plot1, turku_plot1), 0, 20, 'Comparison_of_DM', ('Lisbon', 'Peking', 'Turku'), 'p')
graf_prf((lisbon_plot1, peking_plot1, turku_plot1), 0, 20, 'Comparison_of_DM', ('Lisbon', 'Peking', 'Turku'), 'r')
graf_prf((lisbon_plot1, peking_plot1, turku_plot1), 0, 20, 'Comparison_of_DM', ('Lisbon', 'Peking', 'Turku'), 'f')
graf_prf((lisbon_plot2, peking_plot2, turku_plot2), 0, 20, 'Comparison_of_PSD', ('Lisbon', 'Peking', 'Turku'), 'p')
graf_prf((lisbon_plot2, peking_plot2, turku_plot2), 0, 20, 'Comparison_of_PSD', ('Lisbon', 'Peking', 'Turku'), 'r')
graf_prf((lisbon_plot2, peking_plot2, turku_plot2), 0, 20, 'Comparison_of_PSD', ('Lisbon', 'Peking', 'Turku'), 'f')
```
