```python
import pygal
```

```python
def singletons_graph(data, name):
    print(data)
    chart = pygal.Line()
    chart.x_labels = [a[0] for a in data]
    chart.title = 'Singletons {}'.format(name)
    
    chart.add('Gold', [a[1] for a in data])
    chart.add('Predicted', [a[2] for a in data])
    chart.add('Match', [a[3] for a in data])
    
    chart.render_to_file('plots/singletons/singletons_match.svg')
```

```python
singletons_graph(lisbon_singletons_pos_dm, 'DM Lisbon')
```
