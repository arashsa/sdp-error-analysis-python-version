```python
import pygal
from pygal.style import CleanStyle
```

```python
def draw_graph_pos_frame_numbers(data, name, f, to):
    chart = pygal.Bar(x_label_rotation=30, logarithmic=True, style=CleanStyle)
    chart.x_labels = [a[0] for a in data][f:to]
#     chart.title = name.replace('_', ' ')
    
    chart.add('Gold', [a[1] for a in data][f:to])
    chart.add('Predicted', [a[2] for a in data][f:to])
    chart.add('Match', [a[3] for a in data][f:to])
    
    chart.render_to_file('plots/pos_frame/{}_frequencies.svg'.format(name.replace('_', ' ')))
```

```python
def draw_graph_pos_p_r_f(data, name, f, to):
    chart = pygal.Bar(x_label_rotation=30, style=CleanStyle)
    chart.x_labels = [a[0] for a in data][f:to]
#     chart.title = name.replace('_', ' ')
    
    chart.add('Precision', [a[4] for a in data][f:to])
    chart.add('Recall', [a[5] for a in data][f:to])
    chart.add('F-Score', [a[6] for a in data][f:to])
    
    chart.render_to_file('plots/pos_frame/{}_p_r_f.svg'.format(name.replace('_', ' ')))
```

```python
t_pos = 20
t_frame = 30

draw_graph_pos_frame_numbers(peking_dm_frame, 'Peking_DM_Frame', 0, t_frame)
draw_graph_pos_p_r_f(peking_dm_frame, 'Peking_DM_Frame', 0, t_frame)

draw_graph_pos_frame_numbers(turku_dm_frame, 'Turku_DM_Frame', 0, t_frame)
draw_graph_pos_p_r_f(turku_dm_frame, 'Turku_DM_Frame', 0, t_frame)

draw_graph_pos_frame_numbers(lisbon_dm_pos, 'Lisbon_DM_PoS', 0, t_pos)
draw_graph_pos_p_r_f(lisbon_dm_pos, 'Lisbon_DM_PoS', 0, t_pos)

draw_graph_pos_frame_numbers(peking_dm_pos, 'Peking_DM_PoS', 0, t_pos)
draw_graph_pos_p_r_f(peking_dm_pos, 'Peking_DM_PoS', 0, t_pos)

draw_graph_pos_frame_numbers(turku_dm_pos, 'Turku_DM_PoS', 0, t_pos)
draw_graph_pos_p_r_f(turku_dm_pos, 'Turku_DM_PoS', 0, t_pos)
```
